```python
# 1 Password Validator
import re
def validate_password(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters long."
    if not any(char.islower() for char in password):
        return False, "Password must contain at least one lowercase letter."
    if not any(char.isupper() for char in password):
        return False, "Password must contain at least one uppercase letter."
    if not any(char.isdigit() for char in password):
        return False, "Password must contain at least one digit."
    if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        return False, "Password must contain at least one special character."

    return True, "Password is valid."

user_password = input("Enter a password to validate: ")
is_valid, message = validate_password(user_password)
print(message)
```

```python
# 2 Area calculator
import math
def calculate_area():
    shape = input("Enter the shape(triangle, square, rectangle, circle: ")
    if shape == "triangle":
        base = float(input("Enter the base of the triangle: "))
        height = float(input("Enter the height of the triangle: "))
        area = 0.5 * base * height
        print(f"The area of the triangle is {area}")
    elif shape == "square":
        side = float(input("Enter the side of the square: "))
        area = side ** 2
        print(f"The area of the square is {area}")
    elif shape == "rectangle":
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        area = length * width
        print(f"The area of the rectangle is {area}")
    elif shape == "circle":
        radius = float(input("Enter the radius of the circle: "))
        return math.pi * (radius ** 2)
        print(f"The area of the circle is {area}")
    else:
        return "Invalid"
calculate_area()
```

```python
# 3 Second Largest Element in Array
def second_largest():
    # Ask the user for input
    user_input = input("Enter a list of numbers: ")
    
    # Convert the input string into a list of integers
    try:
        numbers = list(map(int, user_input.split()))
        if len(numbers) < 2:
            print("You need to enter at least two numbers.")
            return
        
        # Remove duplicates and sort the numbers
        unique_numbers = list(set(numbers))
        unique_numbers.sort()

        # Check if there's a second largest number
        if len(unique_numbers) < 2:
            print("Not enough unique elements to find the second largest.")
        else:
            print(f"The second largest number is: {unique_numbers[-2]}")
    except ValueError:
        print("Invalid input! Please enter a list of numbers.")

# Call the function
second_largest()

```

```python
# 4 Word frequency 
def word_frequency():
    i = input("Enter a list of words: ")
    words = i.split()

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    print("Word Frequency:")
    for word, count in freq.items():
        print(f"{word}: {count}")

word_frequency()
```

```python
# 5 Common Elements Squared
def common_elements_squared():
    # Get input for the first set
    set1_input = input("Enter the first set of integers: ")
    set1 = set(map(int, set1_input.split()))
    
    # Get input for the second set
    set2_input = input("Enter the second set of integers: ")
    set2 = set(map(int, set2_input.split()))
    
    # Find common elements and square them
    common = set1 & set2
    result = {x ** 2 for x in common}
    
    print(f"The squared common elements are: {result}")

# Call the function
common_elements_squared()
```

```python
# 6 Rotate 2D Matrix by 90 Degrees Clockwise
def rotate_matrix():
    n = int(input("Enter the size of the square matrix: "))
    print("Enter the matrix row:")
    matrix = []
    
    for _ in range(n):
        row = list(map(int, input().split()))
        matrix.append(row)
    
    for i in range(n):
        for j in range(i, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
    for row in matrix:
        row.reverse()
    
    print("Rotated Matrix:")
    for row in matrix:
        print(" ".join(map(str, row)))

rotate_matrix()
```

```python
# 7 Longest Increasing Subsequence
def longest_increasing_subsequence():
    user_input = input("Enter a list of integers: ")
    nums = list(map(int, user_input.split()))
    
    n = len(nums)
    longest, start = 0, 0
    current_length = 1
    
    for i in range(1, n):
        if nums[i] > nums[i - 1]:
            current_length += 1
            if current_length > longest:
                longest = current_length
                start = i - current_length + 1
        else:
            current_length = 1
    
    subsequence = nums[start:start + longest]
    print(f"Length of the longest increasing continuous subsequence: {longest}")
    print(f"The subsequence is: {subsequence}")

# Call the function
longest_increasing_subsequence()

```
