```python
from collections import deque

def bfs(graph, start, goal):
    queue = deque([start])
    visited = set() 
    parent = {start: None}  
    while queue:
        current = queue.popleft()
        visited.add(current)

        if current == goal:  
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1], len(visited)

        for neighbor in graph[current]:
            if neighbor not in visited and neighbor not in queue:
                queue.append(neighbor)
                parent[neighbor] = current

    return None, len(visited)  

# Graph representation of the maze 
graph = {
    'A': ['B'],
    'B': ['A', 'C'],
    'C': ['B', 'D', 'G'],
    'D': ['C', 'E'],
    'E': ['D', 'F'],
    'F': ['E'],
    'G': ['C', 'H'],
    'H': ['G', 'K', 'I'],
    'I': ['H', 'J'],
    'J': ['I'],
    'K': ['H', 'L'],
    'L': ['K', 'M'],
    'M': ['L', 'N'],
    'N': ['M']
}

# Run BFS
solution_path, nodes_explored = bfs(graph, 'A', 'N')
print("BFS Solution Path:", solution_path)
print("Nodes Explored by BFS:", nodes_explored)
```

```python
from collections import deque

# Function to perform BFS
def bfs(maze, start, end):
    queue = deque([[start]])  # Use a queue for BFS
    visited = set()           # Keep track of visited nodes

    while queue:
        path = queue.popleft()  # Pop the first path from the queue
        x, y = path[-1]         # Current position is the last element in the path

        if (x, y) == end:
            return path  # Return the path if the end is reached

        if (x, y) not in visited:
            visited.add((x, y))  # Mark the current position as visited

            for dx, dy in [(1, 0), (0, 1), (-1, 0), (0, -1)]:  # Possible movements
                next_x, next_y = x + dx, y + dy
                if 0 <= next_x < len(maze) and 0 <= next_y < len(maze[0]) and \
                        maze[next_x][next_y] != '#' and (next_x, next_y) not in visited:
                    new_path = list(path)
                    new_path.append((next_x, next_y))
                    queue.append(new_path)  # Add the new path to the queue
    return None  # Return None if no path is found

# Function to visualize the maze
def print_maze(maze, path=None):
    for i, row in enumerate(maze):
        for j, cell in enumerate(row):
            if path and (i, j) in path:
                print('*', end=' ')  # Mark the path with '*'
            else:
                print(cell, end=' ')
        print()

# Example maze
maze = [
    ['#', '#', '#', '#', '#', '#', '#'],
    ['#', 'S', ' ', ' ', ' ', ' ', '#'],
    ['#', ' ', '#', ' ', '#', ' ', '#'],
    ['#', ' ', '#', ' ', '#', ' ', '#'],
    ['#', ' ', ' ', ' ', ' ', 'E', '#'],
    ['#', '#', '#', '#', '#', '#', '#']
]

start = (1, 1)  # Start position (S)
end = (4, 5)    # End position (E)

# Solve the maze using BFS
path = bfs(maze, start, end)

print("Initial Maze:")
print_maze(maze)
print("\nBFS Path:", path)

if path:
    # Update the maze to visualize the path
    solved_maze = [row[:] for row in maze]  # Copy the maze
    for x, y in path[1:-1]:  # Mark the path excluding start and end
        solved_maze[x][y] = '*'
    print("\nSolved Maze with Path:")
    print_maze(solved_maze)
```

```python
from queue import Queue

def bfs(maze, start, end):
    queue = Queue()
    queue.put([start])  # Enqueue the start position

    while not queue.empty():
        path = queue.get()  # Dequeue the path
        x, y = path[-1]     # Current position is the last element of the path

        if (x, y) == end:
            return path  # Return the path if end is reached

        for dx, dy in [(1,0), (0,1), (-1,0), (0,-1)]:  # Possible movements
            next_x, next_y = x + dx, y + dy
            if maze[next_x][next_y] != '#' and (next_x, next_y) not in path:
                new_path = list(path)
                new_path.append((next_x, next_y))
                queue.put(new_path)  # Enqueue the new path

# Example usage
maze = [
    ['#', '#', '#', '#', '#', '#'],
    ['#', 'S', ' ', ' ', ' ', '#'],
    ['#', ' ', '#', ' ', '#', '#'],
    ['#', ' ', '#', ' ', ' ', '#'],
    ['#', ' ', ' ', ' ', 'E', '#'],
    ['#', '#', '#', '#', '#', '#']
]
start = (1, 1)  # Start position (S)
end = (4, 4)    # End position (E)
path = bfs(maze, start, end)
print(path)
```
