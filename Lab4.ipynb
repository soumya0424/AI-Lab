```python
# Remove Duplicates from a List
def remove_duplicates(lst):
    seen = []
    result = []
    for item in lst:
        if item not in seen:
            seen.append(item)
            result.append(item)
    return result

lst = list(map(int, input("Enter a list of numbers: ").split()))
print("List after removing duplicates:", remove_duplicates(lst))
```

```python
# Rotate a List to the Right by k Positions
def rotate_list(lst, k):
    k %= len(lst) 
    return lst[-k:] + lst[:-k]

lst = list(map(int, input("Enter a list of numbers: ").split()))
k = int(input("Enter the number of positions to rotate: "))
print("List after rotation:", rotate_list(lst, k))
```

```python
# Average Score for Each Student
def calculate_average(scores_dict):
    return {student: sum(scores) / len(scores) for student, scores in scores_dict.items()}

n = int(input("Enter the number of students: "))
scores_dict = {}
for _ in range(n):
    name = input("Enter the student's name: ")
    scores = list(map(int, input(f"Enter the scores for {name}: ").split()))
    scores_dict[name] = scores

print("Average scores:", calculate_average(scores_dict))
```

```python
# Character Frequency in a String
def char_frequency(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    return freq

s = input("Enter a string: ")
print("Character frequency:", char_frequency(s))
```

```python
# First Non-Repeating Character
def first_non_repeating(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    for char in s:
        if freq[char] == 1:
            return char
    return "None"

s = input("Enter a string: ")
print("First non-repeating character:", first_non_repeating(s))
```

```python
# Top 3 Employees with Highest Salaries
def top_3_employees(salaries):
    sorted_employees = sorted(salaries.items(), key=lambda x: (-x[1], x[0]))
    return sorted_employees[:3]

n = int(input("Enter the number of employees: "))
salaries = {}
for _ in range(n):
    name = input("Enter the employee's name: ")
    salary = int(input(f"Enter the salary for {name}: "))
    salaries[name] = salary

print("Top 3 employees:", top_3_employees(salaries))
```

```python
# Longest Substring with Unique Characters
def longest_unique_substring(s):
    seen = {}
    start = 0
    max_length = 0
    max_substring = ""

    for end, char in enumerate(s):
        if char in seen and seen[char] >= start:
            start = seen[char] + 1
        seen[char] = end
        current_length = end - start + 1
        if current_length > max_length:
            max_length = current_length
            max_substring = s[start:end + 1]

    return max_substring, max_length

s = input("Enter a string: ")
substring, length = longest_unique_substring(s)
print("Longest unique substring:", substring)
print("Length of longest unique substring:", length)
```
