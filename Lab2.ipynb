```python
# Program 1: Add two numbers with type conversion
num1 = input("Enter the first number: ")
num2 = input("Enter the second number: ")
try:
    num1 = float(num1)
    num2 = float(num2)
    result = num1 + num2
    print(f"The sum of the numbers is: {result}")
except ValueError:
    print("Invalid input. Please enter numeric values.")
```

```python
# Program 2: Calculate area of a rectangle and compare with square
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
rectangle_area = length * width
square_side = width / 2
square_area = square_side ** 2
print(f"The area of the rectangle is: {rectangle_area}")
print(f"The area of the square is: {square_area}")
if rectangle_area > square_area:
    print("The rectangle has a larger area.")
else:
    print("The square has a larger area.")
```

```python
# Program 3: Check if an integer is even or odd
num = int(input("Enter an integer: "))
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

```python
# Program 4: Advanced calculator
import math
num1 = float(input("Enter the 1st number: "))
num2 = float(input("Enter the 2nd number: "))
option = input("Enter what do calculation you want to perform: + - * / :")
if(option == "+"):
   print(f"The result is:{num1+num2}")
elif(option == "-"):
   print(f"The result is:{num1-num2}")  
elif(option == "*"):
   print(f"The result is:{num1*num2}")
elif(option == "/"):
   print(f"The result is:{num1/num2}")
else:
   print("Invalid input")
```

```python
# Program 5: Fibonacci series up to a specified number of terms
n = int(input("Enter the number of terms for the Fibonacci series: "))
a, b = 0, 1
print(a,b,end=' ')
count = 0
while(count < n-2):
    c = a + b
    print(c,end=' ')
    a = b
    b = c
    count = count + 1
```

```python
# Program 6: Check if a number is prime
num = int(input("Enter the number: "))
flag = False
if num == 0 or num == 1:
    print(num, "is not a prime number")
elif num > 1:
    for i in range(2,num):
        if(num % i) == 0:
            if(num % i) == 0:
                flag = True
            break
    if flag:
        print(num, "is not a prime number")
    else:
        print(num, "is a prime number")
```
